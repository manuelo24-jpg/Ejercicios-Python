```python
"""Con tiempos = [1.2, 1.5, 1.7, 2.0, 1.3, 1.4, 1.6, 2.1]:
1. Calcula media, mediana y moda.
2. Indica si existe moda real. """

import numpy as np
from scipy import stats
import statistics

times = [1.2, 1.5, 1.7, 2.0, 1.3, 1.4, 1.6, 2.1]
print("La media es {}".format(np.average(times)))
print("La mediana es {}".format(np.mean(times)))
print("La moda es {}".format(stats.mode(times)[0]))
print("La moda corregida es {}".format(statistics.multimode(times)))
```

```python
"""Con visualizaciones = [120, 130, 150, 160, 200, 210, 300, 320]:
1. Calcula Q1, Q2, Q3.
2. Calcula el IQR.
3. Detecta outliers.
4. Representa un boxplot 
"""

import numpy as np
import matplotlib.pyplot as plt

views = [120, 130, 150, 160, 200, 210, 300, 320]
outliersList = []

print("Cuartiles: {}".format(np.percentile(views,[25, 50, 75])))

q1 = float(np.percentile(views,[25]))
q3 = float(np.percentile(views,[75]))
iqr = q3-q1

print("El rango intercuartilico es {}".format(iqr))

for view in views:
    if((float(view) < q1-1.5 * iqr) | (float(view) > q3 + 1.5 * float(iqr))):
        outliersList.append(view)

print("Los valores atipicos son {} ".format(outliersList))

plt.boxplot(views)
plt.show()

```

```python
"""4. Comparación A/B
Sean
A = [30, 35, 50, 55, 40, 45, 60] B = [20, 25, 30, 35, 28, 32, 34]
1. Media, desviación típica y CV.
2. ¿Qué grupo es más constante? el grupo B
3. Representa gráficamente
"""
import numpy as np
import matplotlib.pyplot as plt

firstList = [30, 35, 50, 55, 40, 45, 60]
secondList = [20, 25, 30, 35, 28, 32, 34]
bothList = [firstList,secondList]

listMean = np.mean(bothList)
print("La media de las listas es {}".format(listMean))
print("La media de la primera lista es {}".format(np.mean(firstList)))
print("La media de la segunda lista es {}".format(np.mean(secondList)))

v1 = np.var(firstList)
v2 = np.var(secondList)

print("La varianza de la lista 1 es {} ".format(v1))
print("La varianza de la lista 2 es {} ".format(v2))

print("La desviacion tipica de la lista 1 es {} ".format(np.sqrt(v1)))
print("La desviacion tipica de la lista 2 es {} ".format(np.sqrt(v2)))

plt.plot(firstList)
plt.plot(secondList)
plt.show()
```

```python
"""6. Estudio de Outliers
Con edades = [25, 26, 27, 25, 28, 29, 120]:
1. Detecta outliers.
2. Elimínalos.
3. Encuentra y justifica un valor adecuado para sustituir los outliers: la media"""

ages = [25, 26, 27, 25, 28, 29, 120]

q1 = float(np.percentile(ages,[25]))
q3 = float(np.percentile(ages,[75]))
iqr = q3-q1
index = 0

for age in ages:
    if((float(age) < q1-1.5*iqr) | (float(age) > q3 + 1.5*float(iqr))):
        ages[index] = (float(np.median(ages)))
    index += 1

print("Lista sin valores atipicos {}".format(ages))
```

```python
"""7. Probabilidad condicionada
Dadas las siguientes probabilidades
P(móvil)=0.40, P(tablet)=0.30, P(PC)=0.30. P(corto|móvil)=0.80, P(corto|tablet)=0.50, P(corto|PC)=0.20.
1. Calcula P(corto) y P(móvil|corto)"""

# ============================================================================
# PROBABILIDAD CONDICIONADA - TEOREMA DE BAYES
# ============================================================================

# Datos del problema
# ------------------
# Probabilidades de cada dispositivo
P_movil = 0.40
P_tablet = 0.30
P_PC = 0.30

# Probabilidades condicionadas: P(corto | dispositivo)
P_corto_dado_movil = 0.80
P_corto_dado_tablet = 0.50
P_corto_dado_PC = 0.20

print("=" * 70)
print("PROBLEMA DE PROBABILIDAD CONDICIONADA")
print("=" * 70)
print("\n📊 DATOS DEL PROBLEMA:")
print("-" * 70)
print(f"P(móvil)  = {P_movil}")
print(f"P(tablet) = {P_tablet}")
print(f"P(PC)     = {P_PC}")
print()
print(f"P(corto|móvil)  = {P_corto_dado_movil}")
print(f"P(corto|tablet) = {P_corto_dado_tablet}")
print(f"P(corto|PC)     = {P_corto_dado_PC}")

# ============================================================================
# 1. CALCULAR P(corto) - TEOREMA DE LA PROBABILIDAD TOTAL
# ============================================================================
# P(corto) = P(corto|móvil)·P(móvil) + P(corto|tablet)·P(tablet) + P(corto|PC)·P(PC)

print("\n" + "=" * 70)
print("1. CÁLCULO DE P(corto) - TEOREMA DE LA PROBABILIDAD TOTAL")
print("=" * 70)

# Contribución de cada dispositivo
contribucion_movil = P_corto_dado_movil * P_movil
contribucion_tablet = P_corto_dado_tablet * P_tablet
contribucion_PC = P_corto_dado_PC * P_PC

print("\n📐 Fórmula:")
print("P(corto) = P(corto|móvil)·P(móvil) + P(corto|tablet)·P(tablet) + P(corto|PC)·P(PC)")

print("\n🔢 Sustituyendo valores:")
print(f"P(corto) = {P_corto_dado_movil} × {P_movil} + {P_corto_dado_tablet} × {P_tablet} + {P_corto_dado_PC} × {P_PC}")

print("\n📊 Contribución de cada dispositivo:")
print(f"  - Móvil:  {P_corto_dado_movil} × {P_movil} = {contribucion_movil:.4f}")
print(f"  - Tablet: {P_corto_dado_tablet} × {P_tablet} = {contribucion_tablet:.4f}")
print(f"  - PC:     {P_corto_dado_PC} × {P_PC} = {contribucion_PC:.4f}")

# Probabilidad total
P_corto = contribucion_movil + contribucion_tablet + contribucion_PC

print(f"\n✅ RESULTADO:")
print(f"P(corto) = {contribucion_movil:.4f} + {contribucion_tablet:.4f} + {contribucion_PC:.4f}")
print(f"P(corto) = {P_corto:.4f} = {P_corto:.2%}")

# ============================================================================
# 2. CALCULAR P(móvil|corto) - TEOREMA DE BAYES
# ============================================================================
# P(móvil|corto) = P(corto|móvil) · P(móvil) / P(corto)

print("\n" + "=" * 70)
print("2. CÁLCULO DE P(móvil|corto) - TEOREMA DE BAYES")
print("=" * 70)

print("\n📐 Fórmula del Teorema de Bayes:")
print("P(móvil|corto) = [P(corto|móvil) × P(móvil)] / P(corto)")

print("\n🔢 Sustituyendo valores:")
print(f"P(móvil|corto) = [{P_corto_dado_movil} × {P_movil}] / {P_corto:.4f}")
print(f"P(móvil|corto) = {contribucion_movil:.4f} / {P_corto:.4f}")

# Calcular probabilidad condicional
P_movil_dado_corto = contribucion_movil / P_corto

print(f"\n✅ RESULTADO:")
print(f"P(móvil|corto) = {P_movil_dado_corto:.4f} = {P_movil_dado_corto:.2%}")

# ============================================================================
# RESUMEN DE RESULTADOS
# ============================================================================
print("\n" + "=" * 70)
print("📋 RESUMEN DE RESULTADOS")
print("=" * 70)
print(f"\n1️⃣  P(corto) = {P_corto:.4f} ({P_corto:.2%})")
print(f"    ➜ La probabilidad de que un usuario vea un video corto es {P_corto:.2%}")
print()
print(f"2️⃣  P(móvil|corto) = {P_movil_dado_corto:.4f} ({P_movil_dado_corto:.2%})")
print(f"    ➜ Si sabemos que un usuario vio un video corto,")
print(f"      la probabilidad de que lo haya visto desde móvil es {P_movil_dado_corto:.2%}")

# ============================================================================
# INTERPRETACIÓN
# ============================================================================
print("\n" + "=" * 70)
print("💡 INTERPRETACIÓN")
print("=" * 70)
print(f"""
Datos iniciales:
- P(móvil) = {P_movil:.0%} de usuarios usan móvil
- P(corto|móvil) = {P_corto_dado_movil:.0%} de usuarios de móvil ven videos cortos

Resultados:
- P(corto) = {P_corto:.2%} → El {P_corto:.1%} de TODOS los usuarios ven videos cortos
- P(móvil|corto) = {P_movil_dado_corto:.2%} → El {P_movil_dado_corto:.1%} de los que ven videos 
  cortos lo hacen desde móvil

🔍 Observación importante:
La probabilidad aumentó de {P_movil:.0%} a {P_movil_dado_corto:.1%} porque los usuarios
de móvil tienen mayor tendencia a ver videos cortos ({P_corto_dado_movil:.0%}) comparado
con otros dispositivos.
""")

print("=" * 70)
```

```python
"""8. Teorema de Bayes
Dadas las siguientes probalidades
P(fraude)=0.02, P(alarma|fraude)=0.95, P(alarma|no fraude)=0.05.
1. Calcula P(fraude|alarma)"""

# ============================================================================
# TEOREMA DE BAYES - PROBLEMA DE DETECCIÓN DE FRAUDE
# ============================================================================

# Datos del problema
# ------------------
P_fraude = 0.02                    # Probabilidad de fraude
P_no_fraude = 1 - P_fraude         # Probabilidad de NO fraude
P_alarma_dado_fraude = 0.95        # Probabilidad de alarma si hay fraude
P_alarma_dado_no_fraude = 0.05     # Probabilidad de alarma si NO hay fraude

print("=" * 80)
print("TEOREMA DE BAYES - DETECCIÓN DE FRAUDE")
print("=" * 80)
print("\n📊 DATOS DEL PROBLEMA:")
print("-" * 80)
print(f"P(fraude)              = {P_fraude} = {P_fraude:.1%}")
print(f"P(no fraude)           = {P_no_fraude} = {P_no_fraude:.1%}")
print()
print(f"P(alarma|fraude)       = {P_alarma_dado_fraude} = {P_alarma_dado_fraude:.1%}")
print(f"P(alarma|no fraude)    = {P_alarma_dado_no_fraude} = {P_alarma_dado_no_fraude:.1%}")

# ============================================================================
# PASO 1: CALCULAR P(alarma) - TEOREMA DE LA PROBABILIDAD TOTAL
# ============================================================================
print("\n" + "=" * 80)
print("PASO 1: CALCULAR P(alarma) - TEOREMA DE LA PROBABILIDAD TOTAL")
print("=" * 80)

print("\n📐 Fórmula:")
print("P(alarma) = P(alarma|fraude)·P(fraude) + P(alarma|no fraude)·P(no fraude)")

# Calcular cada componente
componente_fraude = P_alarma_dado_fraude * P_fraude
componente_no_fraude = P_alarma_dado_no_fraude * P_no_fraude

print("\n🔢 Sustituyendo valores:")
print(f"P(alarma) = {P_alarma_dado_fraude} × {P_fraude} + {P_alarma_dado_no_fraude} × {P_no_fraude}")

print("\n📊 Desglose de componentes:")
print(f"  - Alarmas por fraude real:    {P_alarma_dado_fraude} × {P_fraude} = {componente_fraude:.4f}")
print(f"  - Alarmas por falso positivo: {P_alarma_dado_no_fraude} × {P_no_fraude} = {componente_no_fraude:.4f}")

# Probabilidad total de alarma
P_alarma = componente_fraude + componente_no_fraude

print(f"\n✅ RESULTADO:")
print(f"P(alarma) = {componente_fraude:.4f} + {componente_no_fraude:.4f}")
print(f"P(alarma) = {P_alarma:.4f} = {P_alarma:.2%}")

# ============================================================================
# PASO 2: CALCULAR P(fraude|alarma) - TEOREMA DE BAYES
# ============================================================================
print("\n" + "=" * 80)
print("PASO 2: CALCULAR P(fraude|alarma) - TEOREMA DE BAYES")
print("=" * 80)

print("\n📐 Fórmula del Teorema de Bayes:")
print("                P(alarma|fraude) × P(fraude)")
print("P(fraude|alarma) = ──────────────────────────────")
print("                        P(alarma)")

print("\n🔢 Sustituyendo valores:")
print(f"                {P_alarma_dado_fraude} × {P_fraude}")
print(f"P(fraude|alarma) = ───────────────────")
print(f"                  {P_alarma:.4f}")
print()
print(f"                {componente_fraude:.4f}")
print(f"P(fraude|alarma) = ───────────")
print(f"                {P_alarma:.4f}")

# Calcular probabilidad condicional usando Teorema de Bayes
P_fraude_dado_alarma = componente_fraude / P_alarma

print(f"\n✅ RESULTADO:")
print(f"P(fraude|alarma) = {P_fraude_dado_alarma:.4f} = {P_fraude_dado_alarma:.2%}")

# ============================================================================
# VISUALIZACIÓN DE RESULTADOS
# ============================================================================
print("\n" + "=" * 80)
print("📊 VISUALIZACIÓN DE LAS PROBABILIDADES")
print("=" * 80)

# Crear barra visual
def crear_barra(valor, longitud=50):
    """Crea una barra visual para representar una probabilidad"""
    lleno = int(valor * longitud)
    vacio = longitud - lleno
    return "█" * lleno + "░" * vacio

print(f"\nP(fraude)              = {P_fraude:.1%}  {crear_barra(P_fraude)}")
print(f"P(alarma)              = {P_alarma:.1%} {crear_barra(P_alarma)}")
print(f"P(fraude|alarma)       = {P_fraude_dado_alarma:.1%} {crear_barra(P_fraude_dado_alarma)}")

# ============================================================================
# ANÁLISIS DETALLADO
# ============================================================================
print("\n" + "=" * 80)
print("📋 RESUMEN Y ANÁLISIS")
print("=" * 80)

print(f"""
🔍 Pregunta: Si salta la alarma, ¿cuál es la probabilidad de que sea fraude real?

📈 Respuesta: P(fraude|alarma) = {P_fraude_dado_alarma:.4f} = {P_fraude_dado_alarma:.2%}

💡 Interpretación:
──────────────────
• Cuando salta la alarma, hay aproximadamente un {P_fraude_dado_alarma:.1%} de probabilidad 
  de que sea un fraude REAL.

• Esto significa que de cada 100 alarmas:
  - {P_fraude_dado_alarma*100:.0f} son fraudes reales (verdaderos positivos)
  - {(1-P_fraude_dado_alarma)*100:.0f} son falsas alarmas (falsos positivos)

🎯 ¿Por qué solo {P_fraude_dado_alarma:.1%} si la alarma es {P_alarma_dado_fraude:.0%} precisa?
────────────────────────────────────────────────────────────────
Porque el fraude es MUY RARO (solo {P_fraude:.0%} de casos):
  • Hay pocas transacciones fraudulentas (2%)
  • Pero hay MUCHAS transacciones normales (98%)
  • El {P_alarma_dado_no_fraude:.0%} de falsas alarmas sobre el 98% normal genera
    muchos falsos positivos
""")

# ============================================================================
# TABLA COMPARATIVA
# ============================================================================
print("\n" + "=" * 80)
print("📊 DESGLOSE DE 10,000 TRANSACCIONES")
print("=" * 80)

total_transacciones = 10000
transacciones_fraude = int(total_transacciones * P_fraude)
transacciones_normales = total_transacciones - transacciones_fraude

alarmas_fraude_real = int(transacciones_fraude * P_alarma_dado_fraude)
alarmas_falso_positivo = int(transacciones_normales * P_alarma_dado_no_fraude)
total_alarmas = alarmas_fraude_real + alarmas_falso_positivo

print(f"""
Total de transacciones: {total_transacciones:,}
├── Fraudes reales:     {transacciones_fraude:,} ({P_fraude:.1%})
│   └── Detectadas:     {alarmas_fraude_real:,} ({P_alarma_dado_fraude:.0%} de {transacciones_fraude})
│
└── Normales:           {transacciones_normales:,} ({P_no_fraude:.1%})
    └── Falsas alarmas: {alarmas_falso_positivo:,} ({P_alarma_dado_no_fraude:.0%} de {transacciones_normales})

Total de ALARMAS: {total_alarmas:,}
├── Fraudes reales:    {alarmas_fraude_real:,} ({alarmas_fraude_real/total_alarmas*100:.1f}%)  ✓
└── Falsas alarmas:    {alarmas_falso_positivo:,} ({alarmas_falso_positivo/total_alarmas*100:.1f}%)  ✗

➜ De cada {total_alarmas} alarmas, solo {alarmas_fraude_real} son fraudes reales
➜ Esto confirma: P(fraude|alarma) ≈ {alarmas_fraude_real}/{total_alarmas} = {alarmas_fraude_real/total_alarmas:.2%}
""")

# ============================================================================
# CONCLUSIONES
# ============================================================================
print("=" * 80)
print("💡 CONCLUSIONES CLAVE")
print("=" * 80)
print(f"""
1. La probabilidad de fraude pasó de {P_fraude:.0%} a {P_fraude_dado_alarma:.1%} al saltar la alarma.
   ➜ La alarma SÍ aumenta significativamente la sospecha de fraude.

2. Sin embargo, solo {P_fraude_dado_alarma:.1%} de las alarmas son fraudes reales.
   ➜ Esto se debe a que el fraude es muy poco común ({P_fraude:.0%}).

3. El sistema tiene:
   ➜ Alta sensibilidad: {P_alarma_dado_fraude:.0%} de detección de fraudes
   ➜ Pero genera muchos falsos positivos debido a la baja prevalencia

4. Aplicación práctica:
   ➜ Cada alarma requiere revisión manual
   ➜ Se espera que {(1-P_fraude_dado_alarma)*100:.0f}% sean falsas alarmas
   ➜ Es el costo de tener alta sensibilidad en eventos raros
""")

print("=" * 80)
```

```python
"""Dadas las siguientes variables y sus correspondientes valores, 
Publicidad: [1000, 1500, 2000, 2500] Ventas: [120, 150, 170, 200] 
1. Ajusta una regresión lineal. 
2. Predice para 3000€."""

import numpy as np

publicidad = np.array([1000, 1500, 2000, 2500])
ventas = np.array([120, 150, 170, 200])

m, b = np.polyfit(publicidad, ventas, 1)

print(f"Pendiente: {m}")
print(f"Intercepto: {b}")
prediccion = m * 3000 + b
print(f"Predicción para 3000€: {prediccion}")
```
