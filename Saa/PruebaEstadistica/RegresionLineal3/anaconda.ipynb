```python
"""1. Carga y exploración de datos
Importa las librerías necesarias (pandas, numpy, matplotlib, sklearn).
Carga el fichero empleados.csv.
Muestra las primeras filas del conjunto de datos.
Comprueba si existen valores nulos o inconsistencias.
"""

import pandas as pd

dfFires = pd.read_csv("fires-all.csv")[0:1000]

dfFires.sort_values(by=['superficie'], ascending=True, inplace=True)

dfFires.head(20)

```

```python
"""3. Preparación de los datos
Separa las variables independientes (X) y la variable dependiente (y). y:salario X: edad, experiencia, formacion, horas_extra
Divide el conjunto de datos en conjunto de entrenamiento y de prueba (por ejemplo, 80 % / 20 %).
Justifica brevemente por qué es necesaria esta división."""

import matplotlib.pyplot as plt

plt.scatter(dfFires['personal'],dfFires['superficie'])
plt.title("Grafico de superficie por heridos")
plt.xlabel("personal implicado")
plt.ylabel("superficie")
plt.show()

```

```python
"""3. Preparación de los datos
Separa las variables independientes (X) y la variable dependiente (y).
Divide el conjunto de datos en conjunto de entrenamiento y de prueba (por ejemplo, 80 % / 20 %).
Justifica brevemente por qué es necesaria esta división."""


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

x = dfFires[['superficie','personal','perdidas','gastos']]
y = dfFires['superficie']

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state = 42
)
```

```python
"""4. Construcción del modelo
Crea un modelo de regresión lineal.
Entrena el modelo con los datos de entrenamiento.
Muestra la ecuación del modelo obtenido."""

model = LinearRegression()
model.fit(x_train, y_train)

pd.DataFrame({
    'Variable' : x.columns,
    'Coeficiente' : model.coef_
})
```

```python
"""5. Evaluación del modelo
Realiza predicciones sobre el conjunto de prueba.
Calcula al menos una métrica de evaluación (R2, MAE o MSE).
Interpreta el resultado obtenido."""

y_pred = model.predict(x_test)

print("R2:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```
